```python
import pandas as pd
```

```python
df = pd.read_excel("data/ICE_ERO_PFD_DataFY2021_FY2025_Q1.xlsx", sheet_name="ICE Removals")
#replace country with 3 digit code; merge criminality column and drop it, 
```

```python
print(df.columns)
df_grouped = df.groupby(['Country of Citizenship', 'Fiscal Year'], as_index=False)['Removals'].sum()
```

```python
print(df_grouped)

import pycountry

def name_to_iso3(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        return None

df_grouped['iso3'] = df_grouped['Country of Citizenship'].str.title().apply(name_to_iso3)
```

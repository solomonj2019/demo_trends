## Subset data

### This file generates only the data we need for our demonstration--i.e. subsets of the existing data

```python
import pandas as pd
```

```python
#Start with the QoL statistics--here we will only use hdi, eys and li for 2023:
df_qol = pd.read_csv("data/cleaned_metrics.csv")
df_qol.head()
```

```python
sa_and_usa = df_qol[
    (df_qol['region'] == 'LAC') | (df_qol['iso3'] == 'USA')
]

cols_2023 = ['iso3', 'country', 'region', 'hdi_2023', 'le_2023', 'eys_2023']
sa_and_usa = sa_and_usa[cols_2023]
```

```python
sa_and_usa.to_csv("data_for_vis/qol.csv")
```

```python
#now, handle the immigration statistics:
df_im = pd.read_csv("data/cleaned_migration.csv")
df_im.columns
```

```python
lac_iso3 = sa_and_usa[sa_and_usa['region'] == 'LAC']['iso3'].tolist()
all_codes = lac_iso3 + ['USA']

migration_subset = df_im[
    (df_im['destination code'].isin(all_codes)) &
    (df_im['origin code'].isin(all_codes))
][['origin code', 'destination code', '2024']]
```

```python
migration_subset
```

```python
migration_subset = df_im[
    (df_im['destination code'] == 'USA') &
    (df_im['origin code'].isin(lac_iso3))
][['origin code', 'destination code', '2024']]
```

```python
migration_subset
```

```python
migration_subset.to_csv("data_for_vis/migration.csv")
```
